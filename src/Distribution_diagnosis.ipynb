```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

df = pd.read_csv("../../data/metadata_(for_scores).csv")

```

```python
if 'diagnostic' in df.columns:
    print("\nDiagnostic Distribution:")
    print(df['diagnostic'].value_counts())
    
    plt.figure(figsize=(12, 6))
    diagnostic_counts = df['diagnostic'].value_counts()
    diagnostic_counts.plot(kind='bar', color='coral')
    plt.title('Distribution of Cancer Types (Diagnostic)', fontsize=14, fontweight='bold')
    plt.xlabel('Diagnostic')
    plt.ylabel('Count')
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    plt.show()
```

```python
if 'diagnostic' in df.columns:
    diagnostic_counts = df['diagnostic'].value_counts()
    total = diagnostic_counts.sum()

    fig, ax = plt.subplots(figsize=(12, 8))

    colors = [
        "#4E79A7", 
        "#F28E2B", 
        "#59A14F", 
        "#E15759", 
        "#B07AA1",
        "#9C755F", 
    ]

    wedges, texts, autotexts = ax.pie(
        diagnostic_counts.values,
        labels=None,
        autopct='%1.1f%%',
        startangle=110,
        counterclock=False,
        pctdistance=0.72,
        colors=colors[:len(diagnostic_counts)],
        wedgeprops={
            'edgecolor': 'white',
            'linewidth': 1.5
        },
        textprops={
            'fontsize': 14,
            'color': 'black',
            'fontweight': 'bold'
        }
    )

    for i, autotext in enumerate(autotexts):
        percentage = diagnostic_counts.values[i] / total * 100

        if percentage < 4:
            autotext.set_fontsize(14)
            autotext.set_bbox({
                'boxstyle': 'round,pad=0.25',
                'facecolor': 'white',
                'edgecolor': 'none',
                'alpha': 0.9
            })

    ax.legend(
        wedges,
        [f'{label}: {count}' for label, count in zip(diagnostic_counts.index, diagnostic_counts.values)],
        title='Diagnostic',
        loc='center left',
        bbox_to_anchor=(1.02, 0.5),
        fontsize=12,
        title_fontsize=11,
        frameon=True
    )

    fig.suptitle(
    'Distribution of Cancer Types (Diagnostic)',
    fontsize=16,
    fontweight='bold',
    x=0.5,
    y=0.95
    )


    ax.axis('equal')
    plt.subplots_adjust(right=0.75, top=0.88)
    plt.show()
```
