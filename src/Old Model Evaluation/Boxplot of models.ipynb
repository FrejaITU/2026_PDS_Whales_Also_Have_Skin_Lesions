```python
import numpy as np
import matplotlib.pyplot as plt

# Your model names
models = ["KNN", "Random Forest", "Decision Tree", "Logistic Regression"]

# Fill in your real values here
means = np.array([0.7685134347531056, 0.7879, 0.731639838466476, 0.7788399295996271])   # example means
stds  = np.array([0.01962566946332609, 0.0332, 0.03152191087278163, 0.022801465715925517])  # example stds

x = np.arange(len(models))

plt.figure(figsize=(8, 5))

plt.errorbar(
    x,
    means,
    yerr=stds,
    fmt='o',
    capsize=6,
    markersize=8,
    color='black',
    ecolor='black',
    linewidth=1.5
)

plt.xticks(x, models, rotation=20)
plt.ylabel("AUC")
plt.title("Model AUC (mean ± std)")
plt.grid(axis='y', linestyle='--', alpha=0.4)

plt.tight_layout()
plt.show()
```
